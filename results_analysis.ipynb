```
context_nsp = []
context_nsp.append(pd.read_csv("results_context_NSP 0.csv"))
context_nsp.append(pd.read_csv("results_context_NSP 1.csv"))
context_nsp.append(pd.read_csv("results_context_NSP 2.csv"))
context_nsp.append(pd.read_csv("results_context_NSP 3.csv"))
context_nsp.append(pd.read_csv("results_context_NSP 4.csv"))

no_context = []
no_context.append(pd.read_csv("results_context_free 0.csv"))
no_context.append(pd.read_csv("results_context_free 1.csv"))
no_context.append(pd.read_csv("results_context_free 2.csv"))
no_context.append(pd.read_csv("results_context_free 3.csv"))
no_context.append(pd.read_csv("results_context_free 4.csv"))

```

```
import numpy as np


def evaluate(runs_list):
  acc = []
  f1 = []
  precision = []
  recall = []
  for data in runs_list:
    acc.append(accuracy_score(data['Label 1'],data['preds']))
    f1.append(f1_score(data['Label 1'],data['preds'],average='macro'))
    precision.append(precision_score(data['Label 1'],data['preds']))
    recall.append(recall_score(data['Label 1'],data['preds']))
  mean = np.array([np.mean(acc), np.mean(f1), np.mean(precision), np.mean(recall)])
  mean = np.round(mean,4)
  std = np.array([np.std(acc), np.std(f1),np.std(precision),np.std(recall)])
#   std = np.format_float_scientific(std, exp_digits=1)
  std = np.round(std,4)
  
  metrics = ["Accuracy","Macro F1","Precision","Recall"]
  print("Means")
  for i in range(4):
    print(metrics[i],mean[i])  
  print("Stds")    
  for i in range(4):
    print(metrics[i],std[i])

```

```
evaluate(context_nsp)
print()
evaluate(no_context)
print()
```
